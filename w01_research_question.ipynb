## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The dataset shows measurable variation in performance across pages.

The average CTR is approximately 0.51, while the average position is around 16.34. This suggests that many pages are not ranking in top positions, yet still achieve varying levels of engagement.

The correlation between avg_position and CTR is -0.07, indicating a weak negative relationship. While better positions tend to have slightly higher CTR, the relationship is not strong, suggesting that other factors also influence performance.

These observations support the need for prioritization, as not all pages respond equally to ranking improvements.

```python
import pandas as pd

df = pd.read_csv("/content_refresh_anonymized.csv")

print("Loaded ✅ (using sample data)")
df.head()
```

```python
print("Average CTR:", round(df["ctr"].mean(), 4))
print("Average Position:", round(df["avg_position"].mean(), 2))

corr = df["avg_position"].corr(df["ctr"])
print("Correlation:", round(corr, 4))
```
